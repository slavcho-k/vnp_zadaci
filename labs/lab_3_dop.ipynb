```python
import pandas as pd
```

```python
gold = pd.read_csv(r'C:\Users\Slavcho\Desktop\1990-2021.csv')
gold = gold[['Date', 'United States(USD)']]
gold['Date'] = pd.to_datetime(gold['Date'])
gold.rename(columns = {'United States(USD)':'Gold Price Close'}, inplace = True)

gold.head()
```

```python
btc = pd.read_csv(r'C:\Users\Slavcho\Desktop\Bitcoin prices.csv')
btc = btc[['Date', 'Close']]
btc['Date'] = pd.to_datetime(btc['Date'])
btc.rename(columns = {'Close':'BTC Close'}, inplace = True)

btc.head()
```

```python
eth = pd.read_csv(r'C:\Users\Slavcho\Desktop\Ethereum prices.csv')
eth = eth[['Date', 'Close']]
eth['Date'] = pd.to_datetime(eth['Date'])
eth.rename(columns = {'Close':'ETH Close'}, inplace = True)

eth.head()
```

```python
df = pd.merge(gold, btc, on='Date')
df = pd.merge(df, eth, on='Date')
df.drop('Date', axis=1, inplace=True)
df
```

```python
df['Shifted'] = df['Gold Price Close'].shift(-3, axis=0)
df
```

```python
from sklearn.model_selection import train_test_split

df_copy = df.copy()

X_train, X_test, Y_train, Y_test = train_test_split(df_copy,  df['Gold Price Close'], test_size=0.2)
```

```python
from xgboost import XGBRegressor

n_estimators_value = 100
max_depth_value = 3
learning_rate_value = 0.1
objective_value = 'reg:squarederror'

xgb_regressor = XGBRegressor(
    n_estimators=n_estimators_value,
    max_depth=max_depth_value,
    learning_rate=learning_rate_value,
    objective=objective_value
)

xgb_regressor.fit(X_train, Y_train)

y_pred = xgb_regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae_xgb = mean_absolute_error(Y_test, y_pred)
mse_xgb = mean_squared_error(Y_test, y_pred)
r2_xgb = r2_score(Y_test, y_pred)

print(f"Mean Absolute Error: {mae_xgb}")
print(f"Mean Squared Error: {mse_xgb}")
print(f"R-squared: {r2_xgb}")
```
