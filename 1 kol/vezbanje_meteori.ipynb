```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/datasets/nasa_asteroids.csv')
df
```

```python
df.isnull().sum()
```

```python
even_df = pd.concat([df[df['Hazardous'] == 0][:755], df[df['Hazardous'] == True]], axis=0)
even_df['Hazardous'].value_counts()
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

iterative_imputer = IterativeImputer(random_state=0)

df_copy = even_df.copy()
df_copy['Absolute Magnitude'] = iterative_imputer.fit_transform(df_copy['Absolute Magnitude'].to_numpy().reshape(-1, 1))
df_copy['Asc Node Longitude'] = iterative_imputer.fit_transform(df_copy['Asc Node Longitude'].to_numpy().reshape(-1, 1))
df_copy['Orbital Period'] = iterative_imputer.fit_transform(df_copy['Orbital Period'].to_numpy().reshape(-1, 1))
df_copy['Perihelion Time'] = iterative_imputer.fit_transform(df_copy['Perihelion Time'].to_numpy().reshape(-1, 1))

df_copy.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split

df_copy.drop('Unnamed: 0.1', axis=1, inplace=True)
df_copy.drop('Unnamed: 0', axis=1, inplace=True)
df_copy.drop('Name', axis=1, inplace=True)
df_copy.drop('Neo Reference ID', axis=1, inplace=True)


X_train, X_test, Y_train, Y_test = train_test_split(df_copy,  even_df['Hazardous'], test_size=0.2)

df_copy
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(X_train, Y_train)
```

```python
from sklearn.metrics import f1_score

Y_pred = classifier.predict(X_test)

print(f1_score(Y_test, Y_pred))
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter= 15000)
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)

print(f1_score(Y_test, Y_pred))
```
