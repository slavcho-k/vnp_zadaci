```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
from IPython.display import HTML
```

```python
import warnings

requests.packages.urllib3.disable_warnings()
warnings.filterwarnings("ignore")
```

```python
url = 'https://www.filmaffinity.com/us/rdcat.php?id=upc_netflix#2022-04-09'
response = requests.get(url)
raw_html = response.text
html = BeautifulSoup(raw_html, "html.parser")
```

```python
movies_html = html.select('.top-movie')
movies = []

for i in range(0, len(movies_html)):
  movie = {}
  movie['Title'] = movies_html[i].select_one('h3').text
  movie['Genre'] = movies_html[i].select_one('.genre').text
  movie['Date'] = movies_html[i].select_one('.date').text
  movies.append(movie)

print(movies)
```

```python
df = pd.DataFrame(movies)
df
```
